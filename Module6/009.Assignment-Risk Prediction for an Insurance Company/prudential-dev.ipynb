# Scale

````python
from sklearn.pipeline import Pipeline

from sklearn.impute import SimpleImputer 
from sklearn.feature_selection import SelectFromModel # feature selection
from sklearn.preprocessing import OneHotEncoder # for categorical
from sklearn.decomposition import PCA # for numerical continuous
from sklearn.preprocessing import StandardScaler # for discrete

from sklearn.compose import ColumnTransformer

from sklearn.model_selection import cross_val_score, StratifiedShuffleSplit

from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
import xgboost as xgb
from xgboost import XGBClassifier
````

````python
# Categorical

cat_pipeline = Pipeline(steps = [
    ('cat_impute', SimpleImputer(strategy='most_frequent')),
    ('cat_one_hot_encoder', OneHotEncoder(sparse_output=True, handle_unknown='ignore', drop='first' ))
]
)
````

````python
# Numerical

num_pipeline = Pipeline(steps = [
    ('num_impute', SimpleImputer(strategy='mean')),
    # ('num_pca', PCA(n_components=.6))
    ('num_scale', StandardScaler())
]
)
````

````python
# Discrete

discrete_pipeline = Pipeline(steps = [
    ('discrete_impute', SimpleImputer(strategy='mean')),
    ('discrete_scale', StandardScaler())
]
)
````

````python
col_transformer = ColumnTransformer(transformers = [
    ('cat_pipeline', cat_pipeline, cat_cols),    
    ('num_pipeline', num_pipeline, num_cols),
    ('discrete_pipeline', discrete_pipeline, discrete_cols)
],
remainder = 'drop',
n_jobs = -1
)
````

````python
# Feature Selection

# estimator = RandomForestClassifier(n_estimators=50, random_state=42, max_depth=5)
# features_selected = SelectFromModel(estimator=estimator, threshold="median")
````

````python
#rfc = RandomForestClassifier(random_state=42, max_depth=5)

xgb_params = {
    'objective': 'binary:logistic',
    'max_depth': 3,
    'learning_rate': 0.1,
}


xgb = XGBClassifier(**xgb_params)
````

````python
def etrees_crossval(n_estimators,
                    max_features,
                    min_samples_split, 
                    min_samples_leaf,
                    max_samples):
    
    pipeline = Pipeline([
        ('preprocessor', col_transformer),
        ('classifier', XGBClassifier(
            n_estimators=int(n_estimators),
            max_features=int(max_features),
            min_samples_split=int(min_samples_split),
            min_samples_leaf=int(min_samples_leaf),
            max_samples=float(max_samples),
            bootstrap=True,
            class_weight='balanced',  # ← Add this for imbalanced classes
            n_jobs=-1,
            random_state=1234,
            verbose=0
        ))
    ])
     
    # Consider using balanced_accuracy instead of regular accuracy
    y_train_inverted = y_train - 1
    mean_cv_score = cross_val_score(pipeline, X_train, y_train_inverted, 
                                    scoring='balanced_accuracy',  # ← Better for imbalanced
                                    cv=5, 
                                    n_jobs=-1).mean()
     
    return mean_cv_score
````

````python
from bayes_opt import BayesianOptimization

#etrees_crossval(10, 3, 2, 1, 1, X_train)

optimizer = BayesianOptimization(
    f=etrees_crossval,
    pbounds={
        'n_estimators': (25, 500),
        'max_features': (1, 20),
        'min_samples_split': (2, 30),
        'min_samples_leaf': (1, 14),
        'max_samples': (0.6, 0.99)
    },
    random_state=1234,
    verbose=2
)

optimizer.maximize(init_points=5, n_iter=25)

print("\nBest parameters:")
print(optimizer.max)
````

````python

pipefinal = make_pipeline(
    col_transformer,
    optimizer
)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

y_train = y_train.values.ravel()
y_train = le.fit_transform(y_train)

pipefinal.fit(X_train, y_train)

print(pipefinal.named_steps)
````

````python
print(cross_val_score(pipefinal, X_train, y_train, cv=5))
````

````python
from sklearn.model_selection import GridSearchCV

params = {'xgbclassifier__n_estimators': [5, 10],
          'xgbclassifier__max_depth': [5, 7]}

gcv = GridSearchCV(pipefinal, param_grid=params, n_jobs=-1, cv=3)
gcv.fit(X_train, y_train)
````

````python
print(gcv.best_params_)
print(gcv.best_score_)
````

````python
pipefinal.set_params(**gcv.best_params_)
pipefinal.fit(X_train, y_train)
````
