```python
# Create a new figure with subplots
fig, axes = plt.subplots(2, 3)  # creates a new, empty figure overwriting the previous fig
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins=50,
                        color="black", alpha=0.5)
fig.subplots_adjust(wspace=0, hspace=0)
```

```python
fig = plt.figure()

ax = fig.add_subplot()

ax.plot(np.random.standard_normal(30).cumsum(), color="black", linestyle="dashed", marker="o")
```

```python
fig = plt.figure()

ax = fig.add_subplot()

data = np.random.standard_normal(30).cumsum()

ax.plot(data, color="orange", linestyle="dotted", label="Default");
ax.plot(data, color="black", linestyle="dashed", drawstyle="steps-post", label="steps-post");
ax.axvspan(3, 6, -2, 5, color="red", alpha=0.3, label="axvspan");
yvalue = np.mean(data)
print(yvalue)
ax.axhline(y=yvalue, color="green", alpha=0.3, label="axvline");
ax.legend();
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(np.random.standard_normal(1000).cumsum());
ax.set_autoscale_on
ax.set_xticks([250, 500, 750]);


ax.text(120, 0, "Hello world!", family='Times New Roman', fontsize=10)
ax.arrow(dx=4, dy=10, x=100, y=0, width=0.5, head_width=2, head_length=5, color='blue', alpha=0.5) #discouraged, use annotate instead
ax.annotate(text="here", xy=(100, 0), xytext=(10, 10),
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=12, color='red', family='Arial')
```
